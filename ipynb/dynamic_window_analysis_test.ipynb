# Dynamic Window Analysis test code

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
df = pd.read_csv("/home/decwest/decwest_workspace/dwpp_test_simulation/data/dynamic_window_pure_pursuit_log.csv")
```

```python
threshold = 10.0
diff = df["timestamp"] - df["timestamp"].shift(1)

# thresholdを超えるところでデータを区切る
gap_points = diff > threshold
split_indices = [0] + list(np.where(gap_points)[0]) + [len(df)]
segments = []
for start, end in zip(split_indices, split_indices[1:]):
    segment = df.iloc[start:end].copy()
    if not segment.empty:
        segments.append(segment)

print(f"Found {len(segments)} segments (threshold={threshold})")
for i, segment in enumerate(segments, 1):
    span = segment["timestamp"].iloc[-1] - segment["timestamp"].iloc[0]
    print(f" segment {i}: rows={len(segment)}, span={span:.3f}s")

# print(f"Mean dt: {np.mean(diff[1:]):.4f} s, Std dt: {np.std(diff[1:]):.4f} s")
# print(f"Max dt: {np.max(diff[1:]):.4f} s, Min dt: {np.min(diff[1:]):.4f} s")
```
